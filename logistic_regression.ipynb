```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import math
from decimal import Decimal, localcontext, getcontext
getcontext().prec = 10
```

```python
image1 = Image.open('bliss.jpg')
image1 = image1.convert('RGB')
image1.thumbnail((50,50))
image1.size
```

```python

image2 = ImageOps.grayscale(image1)
image2
```

```python
def model(grayscale, w_0, w_1, w_2):
    red = sigmoid_function(w_0, grayscale)
    green = sigmoid_function(w_1, grayscale)
    blue = sigmoid_function(w_2, grayscale)
    
    #red = red*Decimal(255)
    #green = green*Decimal(255)
    #blue = blue*Decimal(255)
    
    return (red, green, blue)
```

```python
def sigmoid_function(theta, x):
    z = theta*x
    h = Decimal(1)/(Decimal(1+math.exp(1))**(Decimal(-z)) + Decimal(.000000000001))
    
    return h
```

```python
# define loss function
# L = 1/N sum_i (hat{y}_i-y_i)^2
def loss_function(w_0, w_1, w_2, grayscale, ground_truth):
    result_red = (model(grayscale, w_0, w_1, w_2)[0] - ground_truth[0]/Decimal(255)) **2
    result_green = (model(grayscale, w_0, w_1, w_2)[1] - ground_truth[1]/Decimal(255)) **2
    result_blue = (model(grayscale, w_0, w_1, w_2)[2] - ground_truth[2]/Decimal(255)) **2
    
    #result_red = result_red.mean()
    #result_green = result_green.mean()
    #result_blue = result_blue.mean()
    
    return (result_red, result_blue, result_green)
```

```python
# gradient decsent to find the right theta0 and theta1
# L_0: dL/dw_0 = 1/N sum_i 2(hat{y}_i-y_i)
def L(w_0, w_1, w_2, grayscale, ground_truth):
    result_red = Decimal(Decimal(2)*Decimal(model(grayscale, w_0, w_1, w_2)[0] - Decimal(ground_truth[0]/Decimal(255)))) * Decimal(chain_rule(w_0, grayscale))
    result_green = Decimal(Decimal(2)*Decimal(model(grayscale, w_0, w_1, w_2)[1] - Decimal(ground_truth[1]/Decimal(255)))) * Decimal(chain_rule(w_1, grayscale))
    result_blue = Decimal(Decimal(2)*Decimal(model(grayscale, w_0, w_1, w_2)[2] - Decimal(ground_truth[2]/Decimal(255)))) * Decimal(chain_rule(w_2, grayscale))
    
    #result_red = result_red.mean()
    #print(result_red)
    #result_green = result_green.mean()
    #result_blue = result_blue.mean()
    
    return (result_red, result_blue, result_green)
```

```python
def chain_rule(grayscale, theta):
    return grayscale*Decimal(math.exp(1))**(-theta*grayscale)/(Decimal(math.exp(1))**(theta*grayscale) + 1)**2
```

```python
# Stochastic gradient descent
# define parameters
num_iter = 100
alpha = Decimal(0.01) # step sz

# for SGD
batch_size = 1


# initialization
var_w0 = Decimal(0.)
var_w1 = Decimal(0.)
var_w2 = Decimal(0.)



for epoch in range(num_iter):
    for i in range(0, 45, batch_size):
        #print(f"at pixel {i}")
        for j in range(0, 25, batch_size):
            #print(f"at pixel {i}")
            #color = [image1.getpixel((i,j)), image1.getpixel((i+1,j)), image1.getpixel((i,j+1))]
            color = image1.getpixel((i,j))
            grayscale = Decimal(image2.getpixel((i,j)))
            #grayscale = [image2.getpixel((i,j)), image2.getpixel((i+1,j)), image2.getpixel((i,j+1))]
            d_w0, d_w1, d_w2 = L(var_w0,var_w1,var_w2,grayscale,color)
            #update variables
            var_w0 = var_w0 - alpha * d_w0
            var_w1 = var_w1 - alpha * d_w1
            var_w2 = var_w2 - alpha * d_w2
            #print(var_w0, var_w1, var_w2)
        
    # logging after each epoch (finishing all batches)
    if epoch % 10 == 0:
        print('iteration %d:' % (epoch))
        #print(var_theta0)
        #print(model)
        print(f'loss:{(loss_function(var_w0, var_w1,var_w2, grayscale, color))}')
        #print('gradient: dL/dtheta0 = %f, dL/dtheta1 = %f' % (d_theta0, d_theta1))
        #y_pred = f(var_theta0, var_theta1, x)
        #plt.figure()
        #plt.scatter(x,y)
        #plt.plot(x, y_pred)
```
